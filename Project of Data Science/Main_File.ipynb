__Importing Libraries & Importing Deep Learning Libraries__

```python
import cv2  # Import OpenCV library for image and video processing
import numpy as np  # Import NumPy library for numerical operations
from keras.models import load_model  # Import load_model function from Keras to load pre-trained models
from keras.preprocessing.image import img_to_array  # Import img_to_array function from Keras to convert images
import pygame  # Import pygame library for audio playback
import time  # Import time module for time-related operations
```

```python
# Load the pre-trained face cascade classifier model
face_classifier = cv2.CascadeClassifier(r"C:\Users\ASUS\Downloads\haarcascade_frontalface_default.xml")

# Load the pre-trained emotion detection model
classifier = load_model(r"C:\Users\ASUS\Downloads\model.h5")

# Define the emotion labels corresponding to model output
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Define paths to music files based on detected expressions
music_paths = {
    'Happy': r"C:\Users\ASUS\OneDrive\Desktop\music\happy.mp3",
    'Sad': r"C:\Users\ASUS\OneDrive\Desktop\music\sad.mp3",
    'Angry': r"C:\Users\ASUS\OneDrive\Desktop\music\angry.mp3",
    'Disgust': r"C:\Users\ASUS\OneDrive\Desktop\music\disgust.mp3",
    'Fear': r"C:\Users\ASUS\OneDrive\Desktop\music\fear.mp3",
    'Surprise': r"C:\Users\ASUS\OneDrive\Desktop\music\surprise.mp3",
    'Neutral': r"C:\Users\ASUS\OneDrive\Desktop\music\neutral.mp3"
}

# Initialize pygame mixer for audio playback
pygame.mixer.init()

# Open the webcam (0 represents the default camera)
cap = cv2.VideoCapture(0)

# Initialize variables to track music playback
music_playing = False
music_start_time = None
music_duration = 15  # Music playback duration in seconds

while True:
    # Capture frame-by-frame from the webcam
    ret, frame = cap.read()
    if not ret:
        break  # Break out of the loop if there's an issue reading the frame

    labels = []  # List to store detected emotion labels
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert frame to grayscale
    faces = face_classifier.detectMultiScale(gray)  # Detect faces in the grayscale frame

    for (x, y, w, h) in faces:
        # Draw a rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)
        roi_gray = gray[y:y + h, x:x + w]  # Extract the region of interest (ROI) - face
        roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)  # Resize the ROI to model input size

        if np.sum([roi_gray]) != 0:
            # Preprocess the ROI for emotion classification
            roi = roi_gray.astype('float') / 255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)

            # Predict the emotion from the preprocessed ROI
            prediction = classifier.predict(roi)[0]
            label = emotion_labels[prediction.argmax()]  # Get the emotion label with the highest probability

            # Display the predicted emotion as text above the face
            label_position = (x, y - 10)
            cv2.putText(frame, label, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Play the corresponding music based on the detected emotion
            if label in music_paths:
                if not music_playing:
                    # Start playing the music if it's not already playing
                    pygame.mixer.music.load(music_paths[label])
                    pygame.mixer.music.play()
                    music_playing = True
                    music_start_time = time.time()  # Record the start time of music playback
                else:
                    # Check if music playback duration has elapsed
                    elapsed_time = time.time() - music_start_time
                    if elapsed_time >= music_duration:
                        # Stop the music playback after the specified duration
                        pygame.mixer.music.stop()
                        music_playing = False

        else:
            cv2.putText(frame, 'No Faces', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the resulting frame with emotion detection
    cv2.imshow('Emotion Detector', frame)

    # Check for key press to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
